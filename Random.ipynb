```python
import random
```

```python
# To prints  random number 
print(random.random())
```

```python
# Choice: To print random number from list

list1 = [5,55,22,33,85,99]
print(random.choice(list1))
print(random.choice(list1))
print(random.choice(list1))
```

```python
#Choices. to print by using weights 

list2 = [1,2,3]
print(random.choices(list2, weights= [5,1,2], k= 10))
#weights 5 times more of one's, two times more of three's
# k defining the length of returned list
```

```python
# To print random number from string
# a is  string
a = '68254258'
print(random.choice(a))
print(random.choice(a))
print(random.choice(a))
```

```python
# to print random number between the range of a, b
a=10
b=20
print(random.randint(a,b))

#or
print(random.randint(50,68))


#randrange(a,b) does same

print(random.randrange(10,20))
```

```python
#uniform returns a random floating number between range
print(random.uniform(10,20))
```

```python
# # random number generates based on the seed value, so if the seed value is 3, you will always get perticular as the first random number.
random.seed(3)
print(random.random())
print(random.random())
print(random.random())
print(random.random())
```

```python
#Given the same seed value 
random.seed(3)
print(random.random())
print(random.random())
print(random.random())
print(random.random())
```

```python
# captures the state
print(random.getstate())
```

```python
#setstate() used to rstore the state of random number
#Once it captures, it returns the same captured number

print('random number 1', random.random())
print('random number 2', random.random())
get_state = random.getstate()
print('random number 3', random.random())
print('random number 4', random.random())
random.setstate(get_state)
print('random number 5', random.random())
print('random number 6', random.random())
print('random number 7', random.random())
```

```python
#random get rand bits() used to return specified bits size integer
print(random.getrandbits(2))
print(random.getrandbits(4))
print(random.getrandbits(8))
print(random.getrandbits(12))
```

```python
#shuffle used to rearrage the list
list3 = ["ab","cd","ef"]

random.shuffle(list3)
print(list3)
random.shuffle(list3)
print(list3)
random.shuffle(list3)
print(list3)
```

```python
#sample returns a list with a randomly selected items from sequence
list4 = [1,4,8,10]
print(random.sample(list4,k=3))
print(random.sample(list4,k=2))
print(random.sample(list4,k=3))
```
